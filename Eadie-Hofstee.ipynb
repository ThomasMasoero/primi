```python
df['1/s'] = 1 / df['s']
df['1/v_norm'] = 1 / df['v_norm']
df['v_norm/s'] = df['v_norm'] / df['s']
df
```

```python
from scipy.stats import linregress
slope, intercept, r_value, p_value, std_err = linregress( df['v_norm/s'], df['1/v_norm'] )
'The slope value is {}, the intercept value  is {}'.format(slope, intercept)
```

```python
from numpy import array 
xmax = df['1/s'].max()        # Estremo destro dei valori della x del grafico
x = array( [0, xmax] )        # Un array di 2 punti tra 0 e xmax
y = x * slope + intercept  
```
