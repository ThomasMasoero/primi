```python
import pandas as pd

df = pd.read_excel (r'C:\Users\thomas\Documents\universit√†\valetti\line.xlsx') 
df.columns = ['s', 'v_norm', 'v_inib', 'conc_inib'] 
msus= input("Enter the measure unit of substrate:")
msuv= input("Enter the measure unit of velocity :")
msui=input("Enter the measure unit of inhibent if present:")
print (df)
print('substrate:'+msus)
print('velocity:'+msuv)
print('inhibent concentration:'+msui)
```

```python
from bokeh.plotting import figure, show, output_notebook
param = dict(width = 700, height = 250,
             tools = 'wheel_zoom, reset,pan, box_zoom',
             tooltips = [( 'substrate',   '@s'), ( 'velocity',   '@v')],
            )
output_notebook()
```

```python
df['s/v_norm'] = df['s'] / df['v_norm']
df
```

```python
p = figure(x_axis_label='Concentrazione substrato '+msus,
           y_axis_label='Velocit√† di reazione '+msuv,
           **param,
           )
p.circle( 's', 'v_norm', source=df ) # Plotto i punti (s,v)
show( p )
```

```python
from scipy.stats import linregress
slope, intercept, r_value, p_value, std_err = linregress( df['s'], df['s/v_norm'] )
'slope {}, intercept {}'.format(slope,intercept)  
```

```python
from numpy import array 
```

```python
#problemi col grafico
xmax = df['s/v_norm'].max()
x = array( [-0.5, 0.5] )            
y = x * slope + intercept 

p = figure(title = 'Hanes', x_axis_label='s', y_axis_label='s/v', **param)
p.circle( 's', 's/v_norm', source=df )
p.line( x, y, color='orange')      
show( p )
Vmax = 1/slope
Km = intercept*Vmax
print('Vmax : '+str(Vmax)+' '+msuv+' Km : '+str(Km)+' '+msus)
```
